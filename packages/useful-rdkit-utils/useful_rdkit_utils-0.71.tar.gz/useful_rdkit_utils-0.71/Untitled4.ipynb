```python
from rdkit import Chem
from rdkit.Chem import rdFingerprintGenerator
from rdkit.DataStructs import TanimotoSimilarity
```

```python
smi_1 = "c1cc(ccc1c2c(n(cn2)CC3CC3)c4ccnc(n4)N)F"
```

```python
Chem.MolFromSmiles(smi_1)
```

```python
smi_2 = "NC1=NC(=CC=C1)C1=C(N=CN1CC1CC1)C1=CC=C(F)C=C1"
```

```python
Chem.MolFromSmiles(smi_2)
```

```python
mol_list = [Chem.MolFromSmiles(x) for x in [smi_1, smi_2]]
```

```python
fpgen = rdFingerprintGenerator.GetMorganGenerator()
```

```python
fp_list = fpgen.GetFingerprints(mol_list)
```

```python
TanimotoSimilarity(fp_list[0],fp_list[1])
```

```python
pat_1 = Chem.MolFromSmarts("c1nc(c(n1)-c1ccccn1)-c1ccccc1")
```

```python
mol_list[1].GetSubstructMatch(pat_1)
```
